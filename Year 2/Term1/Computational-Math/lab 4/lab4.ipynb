```python
'''
Computational Mathematics
Lab â„–4
# 1

Author: Eugene Lavrinovych
'''


import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import derivative


def to_show(arr):
    for i in range(len(arr)):
        arr[i] = round(arr[i], 2)
    arr = '\n'.join(map(str, arr))
    return arr


def f(x):
    return x**5 - 4*x**3 + (1.5*x) / (1+x**2)


def bisection(a, b, e=1e-8):
    if np.sign(f(a)) == np.sign(f(b)):
        raise ValueError('f(a) and f(b) must have different signs')
    lst_e = []
    f_a = f(a)
    for i in range(1, 1000):
        lst_e.append(abs(b-a))
        x = a + (b - a)/2
        f_x = f(x)
        print(f'i: {i}    a: {a:.8f}    b: {b:.8f}    |b-a|: {abs(b-a):.8f}    f(x): {f_x:.8f}')
        if np.sign(f_a) == np.sign(f_x):
            a = x
        else:
            b = x
        if abs(b-a) > e:
            continue
        else:
            x = a + (b-a)/2
            print(f'x: {x:.8f}')
            return x, lst_e
        

def chord(a, b, e=1e-8):
    if np.sign(f(a)) == np.sign(f(b)):
        raise ValueError('f(a) and f(b) must have different signs')
    lst_e = []
    f_a = f(a)
    f_b = f(b)
    x = a
    for i in range(1, 1000):
        lst_e.append(abs(b - a))
        x_prev = x
        x = a - f(a)*((b-a) / (f(b) - f(a)))
        f_x = f(x)
        print(f'i: {i}    a: {a:.8f}    b: {b:.8f}    |b-a|: {abs(b - a):.8f}    f(x): {f_x:.8f}'
              f'    |x_prev - x|: {abs(x_prev - x):.8f}')
        if np.sign(f_a) == np.sign(f_x):
            a = x
            f_a = f_x
        else:
            b = x
            f_b = f_x
        if not abs(x_prev-x) < e:
            continue
        else:
            print(f'x: {x:.8f}')
            return x, lst_e
```

```python
# Lab 3

x = np.arange(-3, 3.1, 0.4)
y = [f(i) for i in x]
F1 = [derivative(f, i) for i in x]
F2 = [derivative(f, i, n=2) for i in x]

plt.title('f(x) = x**5 - 4*x**3 + (1.5*x) / (1+x**2)')
plt.plot(x, y)
plt.axis([-3, 3, -8, 10])

plt.text(4, -5, f'x\n{to_show(x)}')
plt.text(5, -5, f'F\n{to_show(y)}')
plt.text(6, -5, f'F`\n{to_show(F1)}')
plt.text(7, -5, f'F``\n{to_show(F2)}')

plt.show()
```

```python
# Lab 4
a = 1
b = 3
e = 1e-4
```

```python
# Bisection method 
bis = bisection(a, b, e)
```

```python
# Chord method
_chord = chord(a, b, e)
```

```python
# Lg graph
fig2, (lgbis, lgchord) = plt.subplots(1, 2)
fig2.tight_layout(pad=3.0)
lgbis.plot(bis[1])
lgbis.set_title('Bisection lg')
lgbis.set_yscale('log')

lgchord.plot(_chord[1])
lgchord.set_title('Chord lg')
lgchord.set_yscale('log')

plt.show()
```
